Perfecto!!! Now we have a full understanding of how our data trully looks like. From the box plot we can get insights like mean, and the quartiles

To get rid of outliers we will use the interquartile range method, where lower limit = IQR *1.5 upper limit = IQR * 1.5

```python
low_quant = tables_basics["runtime_minutes"].quantile(0.25)
upp_quant = tables_basics["runtime_minutes"].quantile(0.75)

IQR = upp_quant - low_quant

low_limit = low_quant - 1.5 * IQR
upp_limit = upp_quant + 1.5 * IQR

tables_basics["runtime_minutes"].clip(lower = low_limit, upper = upp_limit, inplace=True)
```
